```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif']=['SimHei']      #用来正常显示中文标签
plt.rcParams['axes.unicode_minus'] = False      #用来正常显示负号

amount = 5
E_ = 5
x = [1,3,amount,7,9]
y = [9,7,E_,3,1]

fig=plt.figure(figsize=(12,10))

#添加垂直直线
# plt.axvline(x=amount,ls="-",c="green")
plt.vlines(amount, 0, 9, colors = "g")

plt.xlim(0,10)
plt.ylim(0,10)
plt.plot(x, y,marker='o',markersize=5,markerfacecolor='k',markevery=[[1,2,3]]) #markevery使用花式索引指定需要标记的点

plt.annotate('A',xy=(x[1],y[1]),xytext=(x[1],y[1]+0.2))
plt.annotate('B',xy=(x[2],y[2]),xytext=(x[2]+0.2,y[2]))
plt.annotate('C',xy=(x[3],y[3]),xytext=(x[3]+0.2,y[3]))

plt.text(amount,y[0]+0.2, 'S',fontsize=12,color='red')
plt.text(x[4]+0.2, y[4], 'D',fontsize=12,color='orange')

plt.plot([0, amount], [y[1], y[1]], color ='green', linewidth=0.5, linestyle="--")
plt.plot([0, amount], [E_, E_], color ='green', linewidth=0.5, linestyle="--")
plt.plot([0, x[3]], [y[3], y[3]], color ='green', linewidth=0.5, linestyle="--")

plt.xticks(x,[])
plt.yticks([y[1],y[2],y[3]],['$E_{A}$','$E^{*}$','$E_{C}$'])

plt.annotate('', 
                xy=(x[0], y[2]), 
                xytext=(x[0], y[1]), 
                arrowprops=dict(facecolor='grey', shrink=0.1, headwidth=8, width=2)
            )
plt.annotate('', 
                xy=(x[0], y[2]), 
                xytext=(x[0], y[3]), 
                arrowprops=dict(facecolor='grey', shrink=0.1, headwidth=8, width=2)
            )

plt.text(x[2]+0.3, y[1], 
        '$E_{A}$点存在着超额供给，\n导致本币价值下跌', 
        fontsize=13,
        bbox=dict(facecolor='red', alpha=0.5))
plt.text(x[3]+0.5, y[3], 
        '$E_{C}$点存在着超额需求，\n导致本币价值上升', 
        fontsize=13,
        bbox=dict(facecolor='red', alpha=0.5))


plt.title("外汇市场上的均衡")
plt.xlabel("国内资产的数量", loc = "right")
plt.ylabel("汇率,$E_{t}$\n(外币/本币)", loc = "top", rotation = 0, fontsize = 9)

ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

plt.show()
```

```python
#根据资产需求理论，国内资产需求量最重要的决定因素是国内资产的相对预期回报率。
#汇率水平降低时，意味着本币价值可能上升，也就是说本币升值（这里的汇率采取了每单位外国货币相当于本国货币数量的计算方式，即本国货币的升值就相当于汇率的上升），
#本币上升(升值)的幅度越大，国内资产的相对预期回报率就越高，根据资产需求理论，国内资产的需求量就会增加。

#本国货币会随着国内实际利率的上升而上升
#如果国内利率上升是由预期通货膨胀率上升引起的，本国货币就会贬值。
```
