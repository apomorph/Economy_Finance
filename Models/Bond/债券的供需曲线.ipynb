```python
#针对1年期贴现发行债券，其回报率等于以到期收益率衡量的利率，即这种债券的预期回报率等于利率。
#而债券的价格与利率负相关，债券的价格降低，利率升高，预期回报率升高，根据投资组合选择理论，在其他经济不变的情况下，债券的预期回报率越高，其需求量也就越大。
#因此债券的需求曲线是向下倾斜的

#低利率意味着通过发行债券来筹资的成本降低，企业更愿意通过发行债券来借入更多的资金
#因而债券的供给曲线是向上倾斜的，即随着债券价格的上升，供给数量也相应增加

#在给定价格上，当人们愿意购买(需求)的数量与人们愿意出售(供给)的数量相等时，就实现了市场均衡。
#市场均衡时的价格成为均衡价格或市场出清价格，对应该价格的利率称为均衡利率或市场出清利率

#在经济周期的扩张阶段，随着财富的增长，债券需求量相应增加，需求曲线向右位移。
#对未来利率较高的预期会降低长期债券的预期回报率，减少债券需求，引起需求曲线向左位移，其他替代性资产预期回报率的上升会减少对债券的需求，导致需求曲线向左位移。
#预期通货膨胀率的上升降低了债券的预期回报率，导致债券需求下降，债券需求曲线向左位移。
#债券风险的上升导致债券需求减少，推动需求曲线向左位移，其他替代性资产风险的上升导致债券需求上升，债券需求曲线向右位移。
#债券流动性的增强导致债券需求上升，需求曲线向右位移，其他替代性资产流动性的增强减少了对债券的需求，引起需求曲线向左位移。

#在经济周期的扩张阶段，债券供给就会增加，供给曲线向右位移
#预期通货膨胀率上升会增加债券供给，推动债券供给曲线向右位移
#政府赤字越大，债券的供给量就越大，推动供给曲线向右位移

#费雪效应：预期通货膨胀率的上升会引起利率的上升

####### 如何理解债券价格与利率负相关？ ########
#1.已出售债券的面值，票面利率是固定不变的，即利息固定。收益率再怎么涨，债券持有人也不会多收到一分钱利息
#  债券票面利率是指债券发行者每一年向投资者支付的利息占票面金额的比率
#2.购买买一年期面值为100元的债券，利率5%,则购买价格为 100/(1+5%) = 95元，相当于债务人向你(债权人)借了95元，债务人到期偿还本息100元，
#  而市场利率是不断变化的，如果市场利率上升，债务人借钱的利率应该升高，应该多付息，
#  然而已售债券利息固定不变，此时你(债权人)获得的利息相对于当前的市场利率条件下的利息就少了，
#  也可以理解为债券买贵了，别人现在花费更低的价格就能买到之前你买的债券。
#3.债券的内在价值主要受市场利率的影响，债券的收益率并不代表债券的价值！！！
#4.国债的收益率上升意味着主权信用的降低!!!
#5.一定要区分持有至到期收益率与持有期间收益率
#  价格下降，到期收益率上升，但是持有收益率是降低的。

```
