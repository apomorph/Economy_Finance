```python
import math
import numpy as np
import matplotlib.pyplot as plt
# import matplotlib.collections as collections

plt.rcParams['font.sans-serif']=['SimHei']      
plt.rcParams['axes.unicode_minus'] = False  

A = [9,3]
B = [11,1]
G = [10,3]
H = [10,1]

C_ = 1.4    #自主性消费支出 1.4万亿美元
I_ = 1.2    #自主性投资支出 1.2万亿美元
G_ = 3      #政府购买   3.0万亿美元
T_ = 3      #税收       3。0万亿美元
NX_ = 1.3   #自主性净出口   1.3万亿美元
f_ = 1      #金融脆弱性
mpc = 0.6   #边际消费倾向   
d = 0.3     #投资对实际利率的响应程度
x = 0.1     #净出口对实际利率的响应程度

#实际利率
r = np.arange(8)

#总产出
Y = (C_ + I_ - d*f_ + G_ + NX_ - mpc * T_) * (1 - mpc)**(-1) - (d + x) * (1 - mpc)**(-1) * r

fig=plt.figure(figsize=(12,10))

plt.plot(Y, r)

plt.fill_between(Y, 0, r, facecolor='#918597', alpha=0.3)
plt.fill_between(Y, r, 10, facecolor='#464547', alpha=0.3)


plt.xlim(5,12)
plt.ylim(0,10)
plt.yticks(np.arange(8))


plt.scatter(A[0],A[1],s=20,color='black')
plt.scatter(B[0],B[1],s=20,color='black')
plt.scatter(G[0],G[1],s=20,color='green')
plt.scatter(H[0],H[1],s=20,color='green')

plt.vlines(A[0], 0, A[1], colors="grey", linestyles="--", linewidth=1)
plt.vlines(G[0], 0, G[1], colors="grey", linestyles="--",linewidth=1)
plt.vlines(B[0], 0, B[1], colors="grey", linestyles="--",linewidth=1)
plt.hlines(A[1], 0, A[0], color="grey", linestyle="--", linewidth=1)
plt.hlines(H[1], 0, H[0], color="grey", linestyle="--", linewidth=1)

plt.text(5.5,7,'IS')
plt.annotate('A', xy=tuple(A))
plt.annotate('B', xy=tuple(B))
plt.annotate('G', xy=tuple(G))
plt.annotate('H', xy=tuple(H))

plt.text(6.5,2,'超额需求区域')
plt.text(10,7.5,'超额供给区域')

plt.annotate('', 
                xy=(A[0], A[1]), 
                xytext=(G[0], G[1]), 
                arrowprops=dict(facecolor='grey', shrink=0.1, headwidth=8, width=2)
            )
plt.annotate('', 
                xy=(B[0], B[1]), 
                xytext=(H[0], H[1]), 
                arrowprops=dict(facecolor='grey', shrink=0.1, headwidth=8, width=2)
            )
plt.text(G[0]-0.5, G[1]+0.5,'商品存在超额供给时，产出减少')
plt.text(H[0]-0.5, H[1]-0.5,'商品存在超额需\n求时，产出增加')

plt.text(6, 9,r'$Y=[\overline{C}+\overline{I}-d\overline{f}+\overline{G}+\overline{NX}-mpc\times\overline{T}]\times\frac{1}{1-mpc}-\frac{d+x}{1-mpc}\times r$')

plt.title("IS曲线")
plt.xlabel("总产出Y(万亿美元)", loc = "right")
plt.ylabel("实际利率\nr(%)", loc = "top", rotation = 0, fontsize = 9)

ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

plt.show()
```
